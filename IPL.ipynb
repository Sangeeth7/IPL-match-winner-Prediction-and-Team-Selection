```python
import requests
import re
import bs4
```

```python
def top_players(team1,team2):
    season=[]
    player_name=[]
    Team=[]
    player_points=[]
    player_matches=[]
    player_wickets=[]
    player_dots=[]
    player_fours=[]
    player_sixes=[]
    player_catches=[]
    player_stumpings=[]

    ipl_url="https://www.iplt20.com/stats/2019/player-points"
    result=requests.get(ipl_url)
    ipl_soup=bs4.BeautifulSoup(result.text,'html.parser')

    for player_names in ipl_soup.find_all('div',class_='top-players__player-name'):
        player_name.append(player_names.text)
        season.append('2019')
   

    for player_team in ipl_soup.find_all('div',class_='top-players__team'):
        team_name=player_team.find('span')
        names_list=(team_name.get('class'))
        Team.append(names_list[2])
    

    for each_player_points in ipl_soup.find_all('td',class_='top-players__pts top-players__padded is-active'):
        player_points.append(each_player_points.text)
    

    for each_player_matches in ipl_soup.find_all('td',class_='top-players__m '):
        player_matches.append(each_player_matches.text)
    

    for each_player_wickets in ipl_soup.find_all('td',class_='top-players__w '):
        player_wickets.append(each_player_wickets.text)
    

    for each_player_dots in ipl_soup.find_all('td',class_='top-players__d '):
        player_dots.append(each_player_dots.text)
    

    for each_player_fours in ipl_soup.find_all('td',class_='top-players__4s '):
        player_fours.append(each_player_fours.text)
    

    for each_player_sixes in ipl_soup.find_all('td',class_='top-players__6s '):
        player_sixes.append(each_player_sixes.text)
    

    for each_player_catches in ipl_soup.find_all('td',class_='top-players__c '):
        player_catches.append(each_player_catches.text)
    

    for each_player_stumps in ipl_soup.find_all('td',class_='top-players__s '):
        player_stumpings.append(each_player_stumps.text)
    
    player_points_2019=pd.DataFrame({'season':season,'Player_name': player_name,'Team':Team,'player_matches': player_matches,'player_wickets':player_wickets
              ,'player_dots':player_dots,'player_fours':player_fours,'player_sixes':player_sixes,
              'player_catches':player_catches,'player_stumpings':player_stumpings,'player_points':player_points
             })
    top_11=player_points_2019[(player_points_2019['Team']==team1) | (player_points_2019['Team']==team2)].reset_index(drop=True)
    return top_11.head(11)
    
    
```
